```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
```

```python
Flask is a micro web framework for Python, which means it provides the basic tools and features needed to build web applications, while keeping the core functionality minimalistic and flexible. It was created by Armin Ronacher and is known for its simplicity and ease of use.

The advantages of using Flask framework include:

1. Lightweight and minimalistic: Flask is designed to be minimalistic and unopinionated, providing just the essentials for web development. This lightweight nature allows developers to have more control and flexibility in designing their applications according to their specific requirements.

2. Easy to learn and use: Flask has a relatively simple and straightforward API, making it easy for developers to get started quickly. It has a gentle learning curve, which is especially beneficial for beginners or developers who want to build small to medium-sized applications without the overhead of a full-stack framework.

3. Flexibility: Flask allows developers to make decisions about the structure and components of their applications. It doesn't impose a strict directory structure or force the use of specific libraries or tools. This flexibility enables developers to choose the best approach based on their project's needs.

4. Modular and extensible: Flask follows a modular design approach, allowing developers to add or remove components as needed. It supports various extensions and third-party libraries, enabling developers to easily integrate additional functionality into their applications.

5. Great for prototyping and small projects: Due to its simplicity and minimalistic nature, Flask is ideal for prototyping or building small to medium-sized projects efficiently. It allows developers to focus on specific features without unnecessary complexity.

6. Well-documented and active community: Flask has extensive documentation and an active community of developers. This makes it easy to find help, tutorials, and examples, ensuring that developers can quickly resolve issues or learn new concepts.
```

```python
Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.
```

```python
from flask import Flask


app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

```python
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook.
```

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def details():
    return '''Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'This is the contact page'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))    
        print(url_for('about'))   
        print(url_for('contact')) 
```
